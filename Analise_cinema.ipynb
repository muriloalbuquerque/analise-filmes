```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Top_1000_IMDb_movies_New_version.csv')
```

```python
df
```

```python
df.info()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Movie Rating'], bins=20, kde=True)
plt.title('Distribuição das Avaliações dos Filmes')
plt.xlabel('Classificação do Filme')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Metascore of movie', y='Gross', data=df)
plt.title('Relação entre Metascore e Receita Bruta')
plt.xlabel('Metascore')
plt.ylabel('Receita Bruta (milhões)')
plt.show()
```
